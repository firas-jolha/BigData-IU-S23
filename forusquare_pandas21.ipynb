```python
def extract_df(from_file, group_set, columns, selected_col, sep = "\t", header = None, chunksize=10**4):
    df = None
    for chunk in pd.read_csv(from_file, sep=sep, chunksize = chunksize, header = header, names = columns):
        selected = chunk[chunk[selected_col].isin(group_set)]
        if df is None:
            df = selected
        else:
            df = pd.concat([df, selected])
    return df

def extract_df2(from_file, group_set, columns, selected_col1, selected_col2, sep = "\t", header = None, chunksize=10**4):
    df = None
    for chunk in pd.read_csv(from_file, sep=sep, chunksize = chunksize, header = header, names = columns):
        selected = chunk[chunk[selected_col1].isin(group_set) & chunk[selected_col2].isin(group_set)]
        if df is None:
            df = selected
        else:
            df = pd.concat([df, selected])
    return df
```
